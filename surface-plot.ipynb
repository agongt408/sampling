```python
%matplotlib widget
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.outer(np.linspace(0, 5, 30), np.ones(30))
Y = X.copy().T # transpose
# Z = 2*(X-Y)*(np.heaviside(X-1,0) - np.heaviside(Y-1,0)) # TV

Z = X/Y + Y/X -2 # RevEnt
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='Blues', edgecolor='none', alpha=0.5)
ax.set(xlabel='a', ylabel='b', zlabel='R(a,b)')
plt.tight_layout()
plt.show()
# plt.savefig('R(a,b)-TV')
```
